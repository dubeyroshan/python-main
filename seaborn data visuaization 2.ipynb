```python
### univariate distrbution for categorica(object)


sns.countplot(x="fuel_system",data=df)

plt.show()
```

```python
 sns.countplot(x="highway_mpg",data=df,palette= ['red', 'green', 'blue', 'orange'])
plt.title("my stats",color='red')
plt.xlabel('avg',loc="right")
plt.ylabel("countts",loc='top')

plt.grid()

plt.show()
```
